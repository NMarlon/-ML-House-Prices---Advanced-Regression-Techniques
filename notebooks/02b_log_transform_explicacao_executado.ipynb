# Comparacao: SalePrice COM e SEM Log Transform

Este notebook mostra visualmente o efeito do log transform na distribuicao dos precos.

```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 4)
plt.rcParams['font.size'] = 12

print('Bibliotecas importadas!')
```

```python
# Carregar dados
train = pd.read_csv('../data/train.csv')
price = train['SalePrice']
price_log = np.log1p(price)

print(f'Total de casas: {len(price)}')
print(f'Preco minimo: R$ {price.min():,.0f}')
print(f'Preco maximo: R$ {price.max():,.0f}')
print(f'Media: R$ {price.mean():,.0f}')
print(f'Mediana: R$ {price.median():,.0f}')
print(f'Skew (assimetria): {price.skew():.4f}')
```

## Comparacao Principal: Histograma

Esquerda: sem log (a maioria espremida no lado esquerdo).
Direita: com log (distribuicao mais simetrica, forma de sino).

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].hist(price, bins=50, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].set_title('SalePrice - SEM Log (Precos Reais)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Preco (R$)')
axes[0].set_ylabel('Numero de Casas')
axes[0].axvline(price.mean(), color='red', linestyle='--', label=f'Media R$ {price.mean():,.0f}')
axes[0].axvline(price.median(), color='orange', linestyle='--', label=f'Mediana R$ {price.median():,.0f}')
axes[0].legend()

axes[1].hist(price_log, bins=50, color='coral', edgecolor='white', alpha=0.8)
axes[1].set_title('SalePrice - COM Log (log(preco))', fontsize=14, fontweight='bold')
axes[1].set_xlabel('log(Preco)')
axes[1].set_ylabel('Numero de Casas')
axes[1].axvline(price_log.mean(), color='red', linestyle='--', label=f'Media {price_log.mean():.2f}')
axes[1].axvline(price_log.median(), color='orange', linestyle='--', label=f'Mediana {price_log.median():.2f}')
axes[1].legend()

plt.tight_layout()
plt.savefig('../reports/histograma_comparacao.png', dpi=150, bbox_inches='tight')

print('Sem log: maioria espremida no lado esquerdo.')
print('Com log: distribuicao mais simetrica (forma de sino).')
```

## QQ-Plot: Teste de Normalidade

O QQ-Plot compara os dados com uma distribuicao normal.
Quanto mais proximo da reta, mais normal e a distribuicao.

- Sem log: pontos nas extremidades desviam muito da reta (caudas pesadas)
- Com log: pontos ficam mais proximos da reta (mais normal)

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

stats.probplot(price, dist='norm', plot=axes[0])
axes[0].set_title('QQ-Plot: SEM Log (fora da reta = nao normal)', fontsize=13, fontweight='bold')
axes[0].get_lines()[0].set(color='steelblue', markersize=3)
axes[0].get_lines()[1].set(color='red', linewidth=2)

stats.probplot(price_log, dist='norm', plot=axes[1])
axes[1].set_title('QQ-Plot: COM Log (proximo da reta = normal)', fontsize=13, fontweight='bold')
axes[1].get_lines()[0].set(color='coral', markersize=3)
axes[1].get_lines()[1].set(color='red', linewidth=2)

plt.tight_layout()
plt.savefig('../reports/qqplot_comparacao.png', dpi=150, bbox_inches='tight')
```

## BoxPlot: Visualizando Outliers

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

bp1 = axes[0].boxplot(price, vert=True, patch_artist=True,
                     boxprops=dict(facecolor='steelblue', alpha=0.7))
axes[0].set_title('BoxPlot: SEM Log', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Preco (R$)')
n_outliers_sem = (price > 340000).sum()
axes[0].annotate(f'Outliers: {n_outliers_sem} casas',
                 xy=(0.5, 0.95), xycoords='axes fraction',
                 ha='center', fontsize=11, color='red')

bp2 = axes[1].boxplot(price_log, vert=True, patch_artist=True,
                     boxprops=dict(facecolor='coral', alpha=0.7))
axes[1].set_title('BoxPlot: COM Log', fontsize=14, fontweight='bold')
axes[1].set_ylabel('log(Preco)')
upper_log = np.log1p(340000)
n_outliers_com = (price_log > upper_log).sum()
axes[1].annotate(f'Outliers: {n_outliers_com} casas',
                 xy=(0.5, 0.95), xycoords='axes fraction',
                 ha='center', fontsize=11, color='red')

plt.tight_layout()
plt.savefig('../reports/boxplot_comparacao.png', dpi=150, bbox_inches='tight')

print(f'Sem log: {n_outliers_sem} outliers visiveis')
print(f'Com log: {n_outliers_com} outliers visiveis (menos extremos)')
```

## Tabela Comparativa

```python
comparison = pd.DataFrame({
    'Metrica': ['Media', 'Mediana', 'Desvio Padrao', 'Skew', 'Min', 'Max'],
    'Sem Log': [
        f'R$ {price.mean():,.0f}',
        f'R$ {price.median():,.0f}',
        f'R$ {price.std():,.0f}',
        f'{price.skew():.4f}',
        f'R$ {price.min():,.0f}',
        f'R$ {price.max():,.0f}',
    ],
    'Com Log': [
        f'{price_log.mean():.4f}',
        f'{price_log.median():.4f}',
        f'{price_log.std():.4f}',
        f'{price_log.skew():.4f}',
        f'{price_log.min():.4f}',
        f'{price_log.max():.4f}',
    ],
    'Interpretacao': [
        'Media puxada por valores altos',
        'Media > Mediana = assimetria',
        'Desvio grande = dados espalhados',
        'Skew > 1 = muito assimetrico',
        '',
        '',
    ],
    ' ': [
        'Mais proxima da mediana',
        'Media ~ Mediana = simetrico',
        'Desvio menor = dados concentrados',
        'Skew ~ 0 = distribuicao normal',
        '',
        '',
    ]
})

print('=' * 70)
print('COMPARACAO: SalePrice SEM vs COM Log Transform')
print('=' * 70)
print(comparison.to_string(index=False))

print('\nSkew: < 0.5 = simetrico | 0.5-1 = moderado | > 1 = assimetrico')
print(f'\nSem log: skew = {price.skew():.4f} (assimetria ALTA)')
print(f'Com log: skew = {price_log.skew():.4f} (aproximadamente normal)')
```

## Exemplo Numerico: Casas Especificas

```python
example_prices = [50000, 100000, 150000, 200000, 300000, 400000, 500000, 750000]

print('Preco Real  ->  log(Preco)')
print('-' * 40)
for p in example_prices:
    log_p = np.log1p(p)
    print(f'R$ {p:>10,.0f}  ->  {log_p:>10.4f}')

print('\nObserve:')
print('  R$ 50.000  -> 10.82 (casa barata)')
print('  R$ 750.000  -> 13.53 (casa cara)')
print('  Diferenca na escala log: so 2.71!')
print('')
print('Sem log: R$ 750k - R$ 50k = R$ 700k de diferenca')
print('Com log: 13.53 - 10.82 = 2.71 de diferenca')
print('O log comprime os valores grandes e expande os pequenos.')
```

## Conclusao

**Por que usar log transform?**

1. Torna a distribuicao mais normal - muitos modelos assumem normalidade
2. Reduz o impacto de outliers - valores extremos ficam menos extremos
3. Lineariza relacoes - se o preco cresce exponencialmente com a area, o log torna linear
4. Estabiliza a variancia - a variancia fica mais constante em todas as faixas

**Quando NAO usar?**

- Quando a distribuicao ja e normal
- Quando voce quer interpretar os valores diretamente
- Modelos baseados em arvores (Random Forest, XGBoost) nao precisam de log
